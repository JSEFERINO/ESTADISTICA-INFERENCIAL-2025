## **7. Tomando mil muestras - Bootstrap**




```python
# **A. Usando la tecnica de BootStrap seleccionaremos 1000 muestras aleatorias y las guardaremos**
muestras1 = np.array([]) #En este espacio guardaremos cada muestra tomada
num_muestras = 10000 # Este será el tamaño de las muestra

for m in range(num_muestras): # Con este procedimiento grantizamos tomar las mil muestras
 muestras1 = np.append(muestras1,
 np.random.choice(df['ASISTENCIA2'],tamano_muestra,replace=True)) # guardamos en muestras las seleccionadas de la poblacion inicial
muestras1 = muestras1.reshape(-1, tamano_muestra)
muestras1.shape, muestras1 # nos muestras el tamaño del archivo y las muestras
# **d. Estamos realizando el muestreo de medias**
plt.hist(muestras1.mean(axis=1), bins=100, color="turquoise") # Histograma de la distribución muestral de la media
plt.xlabel("\nCalificaciones promedio de las muestras")
plt.ylabel("Frecuencia\n")
plt.show()
sns.histplot(muestras1.mean(axis=1), kde=True, stat="density", kde_kws=dict(cut=3),
alpha=.4, edgecolor=(1, 1, 1, .4),)

```

## **17. Metodos de Regresión Lineal - _optimize.curve_fit_**




```python
# @title **A. Empecemos la Regresión Lineal - `optimize.curve_fit`**
datax = df3['ESTATURA']
datay = df3['PESO']
def f( x, p0, p1):
  return p0 + p1*x
def ff(x, p):
  return f(x, *p)
# Estos son los verdaderos parámetros
p0 = 1.0
p1 = 1.0
# Estas son conjeturas iniciales para ajustes:
pstart = [
p0 + random.random(),
p1 + random.random(),
]
```

```python
# @title **B. Optimizar la Curva de ajuste - `optimize.curve_fit`**
from scipy import optimize
import numpy as np

err_stdev = 0.2

# Definimos los parametros de nuestra función
def fit_curvefit(p0, datax, datay, function, yerr=err_stdev, **kwargs):
    # usamos el metodo de curva fit
    pfit3, pcov = optimize.curve_fit(function, datax, datay, p0=p0, sigma=None, epsfcn=0.0001, **kwargs)

    # ajustamos los datos y residuos
    error = []
    for i in range(len(pfit3)):
        try:
            # Calculate the standard deviation of the error from the covariance matrix diagonal
            error.append(np.sqrt(np.absolute(pcov[i][i])))
        except:
            error.append(0.00)

    # These variables are local to the function and will be returned
    pfit_curvefit = pfit3
    perr_curvefit = np.array(error)

    return pfit_curvefit, perr_curvefit

# Call the function and store the returned values
pfit_curvefit_result, perr_curvefit_result = fit_curvefit(pstart, datax, datay, f) # Pass 'f' as the function

print('Algunas visualizaciones de lo construido hasta ahora:')
# Use the stored variables outside the function
print('Lo almacenado en pfit_curvefit es ', pfit_curvefit_result)
print('Lo almacenado en perr_curvefit =', perr_curvefit_result)
```

```python
# @title **C. Bondad de Ajuste $R^2$ para el método de regresión `curve.fit`**
from sklearn.metrics import r2_score

# Use the variable that stores the fitted parameters returned by the function
r2 = r2_score(datay, f(datax, *pfit_curvefit_result))

print('El coeficiente de Bondad de Ajuste o de Determinación del modelo R_cuadrado = ', round(r2,3))
print('Esto es, el modelo explica la variabilidad observada en la respuesta en un porcentaje de ', round(r2*100,1), '%')
```

## **20. Regresión Lineal con Bootstrap y Diagrama de Dispersión**




```python
# @title **Paso A. Regresión Lineal con Bootstrap y Diagrama de Dispersión**

# Importar librerías necesarias (asegurarse de que están instaladas)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import optimize

# Definir la función para la regresión lineal
def linear_model(x, a, b):
    return a + b * x

# Número de remuestreos bootstrap
n_bootstrap = 10000

# Listas para almacenar los coeficientes estimados de cada remuestreo
bootstrap_a = []
bootstrap_b = []

# Verificar que datax y datay estén definidos y tengan la misma longitud
assert len(datax) == len(datay), "Los datos de entrada deben tener la misma longitud"

# Realizar remuestreo bootstrap
for _ in range(n_bootstrap):
    # Muestrear con reemplazo de los datos originales
    sample_indices = np.random.choice(len(datax), size=len(datax), replace=True)
    bootstrap_x = datax.iloc[sample_indices] if isinstance(datax, pd.Series) else datax[sample_indices]
    bootstrap_y = datay.iloc[sample_indices] if isinstance(datay, pd.Series) else datay[sample_indices]

    # Ajustar el modelo lineal a la muestra bootstrap
    try:
        popt, _ = optimize.curve_fit(linear_model, bootstrap_x, bootstrap_y)
        bootstrap_a.append(popt[0])
        bootstrap_b.append(popt[1])
    except RuntimeError:
        # Manejar casos donde curve_fit no converge
        continue

# Calcular los coeficientes promedio de bootstrap
mean_a = np.mean(bootstrap_a)
mean_b = np.mean(bootstrap_b)

print(f"Coeficiente 'a' (Intercepto) promedio con Bootstrap: {mean_a:.4f}")
print(f"Coeficiente 'b' (Pendiente) promedio con Bootstrap: {mean_b:.4f}")

# Dibujar el diagrama de dispersión y la línea de regresión bootstrap
plt.figure(figsize=(8, 6))
plt.scatter(datax, datay, label="Datos", alpha=0.6)

# Dibujar la línea de regresión basada en los coeficientes promedio de bootstrap
x_plot = np.linspace(np.min(datax), np.max(datax), 100)
plt.plot(x_plot, linear_model(x_plot, mean_a, mean_b), color="purple",
         label=f"Regresión Bootstrap\ny = {mean_b:.2f}x + {mean_a:.2f}")

plt.xlabel("ESTATURA")
plt.ylabel("PESO")
plt.title("Diagrama de Dispersión y Regresión Lineal con Bootstrap")
plt.legend()
plt.grid(True)
plt.show()

# Visualizar la distribución de los coeficientes bootstrap
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(bootstrap_a, kde=True, color='skyblue')
plt.xlabel("Coeficiente 'a' (Intercepto)")
plt.ylabel("Frecuencia")
plt.title("Distribución Bootstrap del Intercepto")

plt.subplot(1, 2, 2)
sns.histplot(bootstrap_b, kde=True, color='salmon')
plt.xlabel("Coeficiente 'b' (Pendiente)")
plt.ylabel("Frecuencia")
plt.title("Distribución Bootstrap de la Pendiente")

plt.tight_layout()
plt.show()
```

## **23. Muestreo de Medias y Varianzas de una Variable**


```python
# @title **A. Muestreo de Medias y Varianzas de una Variable**
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
n = 1000
a = 18 # Valor minimo en un dado
b = 22 # Valor máximo en un dado
x=1 # nos va indicar el primer lanzamiento
E1 = [] # Edad del Estudiante 1
E2 = [] # Edad del Estudiante 2
E3 = [] # Edad del Estudiante 3
E4 = [] # Edad del Estudiante 4
E5 = [] # Edad del Estudiante 5

suma12 = [] # Edad Total de los cinco estudiantes
media = [] # Edad Promedio de los cinco estudiantes
varianza = [] # Varianza Muestral de la Edad de los cinco estudiantes
suma = 0 # Acumular los resultados sumandolos

while x<=n: # Esto es para decir que queremos lanzar el dado 10 veces
    valor1=random.randint(a,b) # Edad Estudiante 1
    E1.append(valor1) # Almacenar Edad Estudiante 1
    valor2=random.randint(a,b) # Edad Estudiante 2
    E2.append(valor2) # Almacenar Edad Estudiante 2
    valor3=random.randint(a,b) # Edad Estudiante 3
    E3.append(valor3) # Almacenar Edad Estudiante 3
    valor4=random.randint(a,b) # Edad Estudiante 4
    E4.append(valor4) # Almacenar Edad Estudiante 4
    valor5=random.randint(a,b) # Edad Estudiante 5
    E5.append(valor5) # Almacenar Edad Estudiante 5
    suma12.append(valor1+valor2+valor3+valor4+valor5)
    # Calculate the mean for the current iteration
    current_mean = (valor1+valor2+valor3+valor4+valor5)/5
    media.append(current_mean)
    # Use the current mean for variance calculation
    varianza.append((valor1**2+valor2**2+valor3**2+valor4**2+valor5**2-5*(current_mean**2))/4)
    suma=suma+valor1+valor2+valor3+valor4+valor5 # Acumulando la suma de los valores obtenidos
    x=x+1 # el ciclo se cierra y comienza
promedio=suma/n #Deseo hallar la media de los resultados de los 10 lanzamientos
EDAD = pd.DataFrame({'E1':E1,'E2':E2,'E3':E3,'E4':E4,'E5':E5,'Total':suma12,'media':media,'varianza':varianza})
EDAD.head()


```

```python
# @title **B. Diagrama de tallo y hoja**
def stem_leaf_plot(data):
    stems = {}
    for value in data:
    # Handle potential floats by converting to integer if needed or appropriate
    # Assuming data is integers or can be treated as integers for this plot
        value_int = int(value)
        stem = value_int // 10
        leaf = value_int % 10
        if stem not in stems:
          stems[stem] = [leaf]
        else:
          stems[stem].append(leaf)
        # Sort the stems and leaves before printing
        sorted_stems = sorted(stems.keys())
    for stem in sorted_stems:
        sorted_leaves = sorted(stems[stem])
        print(f"{stem} | {' '.join(map(str, sorted_leaves))}")

print(f"Diagram de tallo y Hojas")
stem_leaf_plot(peso)

```

```python

# @title **D. Crear los intervalos**
num_intervalos = round(1+3.322*np.log10(len(peso)),0)
intervalos = []
inicio_intervalo = min(peso)
# Ensure ancho_intervalo is defined before use
rango = max(peso) - min(peso)
ancho_intervalo = round(rango / num_intervalos,1)

for i in range(int(num_intervalos)):
  fin_intervalo = inicio_intervalo + ancho_intervalo
  # Convert the Series to scalar values using .item() before storing
  # Ensure that we handle the last interval to include the maximum value
  if i == int(num_intervalos) - 1:
      intervalos.append((round(inicio_intervalo, 1), round(max(peso), 1)))
  else:
      intervalos.append((round(inicio_intervalo, 1), round(fin_intervalo, 1)))
  inicio_intervalo = fin_intervalo

# Inicializar las listas para almacenar la información de la tabla
marca_clase = []
frecuencia_absoluta = []
frecuencia_absoluta_acumulada = []
frecuencia_relativa = []
frecuencia_relativa_acumulada = []
frecuencia_acumulada = 0

for intervalo in intervalos:
  # Calcular la marca de clase - Now interval elements are scalars
  marca = (intervalo[0] + intervalo[1]) / 2
  marca_clase.append(marca)

  # Contar la frecuencia absoluta del intervalo - Comparison with scalar values from interval tuple
  # Use the correct comparison for the last interval to include the upper bound
  if intervalo == intervalos[-1]: # Check if it's the last interval
      frecuencia = sum(1 for peso_i in peso if intervalo[0] <= peso_i <= intervalo[1])
  else:
      frecuencia = sum(1 for peso_i in peso if intervalo[0] <= peso_i < intervalo[1])

  frecuencia_absoluta.append(frecuencia)
  # Calcular la frecuencia absoluta acumulada
  frecuencia_acumulada += frecuencia
  frecuencia_absoluta_acumulada.append(frecuencia_acumulada)
  # Calcular la frecuencia relativa
  frecuencia_relativa.append(frecuencia / len(peso))
  # Calcular la frecuencia relativa acumulada
  frecuencia_relativa_acumulada.append(frecuencia_acumulada / len(peso))

# Crear un DataFrame de Pandas con la tabla de frecuencias
# Ensure marca_clase elements are scalars before creating Series
tabla_frecuencias = pd.DataFrame({
    'Intervalo': intervalos,
    # Ensure Marca de Clase column is created from scalar values
    'Marca de Clase': marca_clase,
    'Fre_Abs': frecuencia_absoluta,
    'Fre_Abs_Acum': pd.Series(frecuencia_absoluta_acumulada).round(3),
    'Frec_Rel': pd.Series(frecuencia_relativa).round(3),
    'Fre_Rel_Acum': pd.Series(frecuencia_relativa_acumulada).round(3)
})
tabla_frecuencias
```

```python

# --- Calcular frecuencias ---
marca_clase = []
frecuencia_absoluta = []
frecuencia_absoluta_acumulada = []
frecuencia_relativa = []
frecuencia_relativa_acumulada = []

frec_acum = 0
for i, intervalo in enumerate(intervalos):
    # Marca de clase
    marca = (intervalo[0] + intervalo[1]) / 2
    marca_clase.append(marca)

    # Frecuencia absoluta
    if i == len(intervalos) - 1:  # Último intervalo
        freq = sum((peso >= intervalo[0]) & (peso <= intervalo[1])) # Incluir el límite superior
    else:
        freq = sum((peso >= intervalo[0]) & (peso < intervalo[1]))

    frecuencia_absoluta.append(freq)

    # Frecuencias acumuladas
    frec_acum += freq
    frecuencia_absoluta_acumulada.append(frec_acum)
    frecuencia_relativa.append(round(freq / len(peso), 3))
    frecuencia_relativa_acumulada.append(round(frec_acum / len(peso), 3))

# Crear un DataFrame de Pandas con la tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Intervalo': [f"[{ini}, {fin}]" if i == len(intervalos) - 1 else f"[{ini}, {fin})" for i, (ini, fin) in enumerate(intervalos)],
    'Marca de Clase': marca_clase,
    'Frec_Abs': frecuencia_absoluta,
    'Frec_Abs_Acum': frecuencia_absoluta_acumulada,
    'Frec_Rel': frecuencia_relativa,
    'Frec_Rel_Acum': frecuencia_relativa_acumulada
})

print("Tabla de Frecuencias Agrupadas (Último intervalo cerrado):")
print(tabla_frecuencias)

# --- Cálculo de medidas estadísticas (con la tabla corregida) ---
total_datos = len(peso)

# 1. Media
media = sum(mc * f for mc, f in zip(tabla_frecuencias['Marca de Clase'], tabla_frecuencias['Frec_Abs'])) / total_datos
print(f"\nMedia Agrupada: {media:.2f}")

# 2. Moda (usando la tabla_frecuencias corregida)
indice_modal = np.argmax(tabla_frecuencias['Frec_Abs'])
L_modal = float(tabla_frecuencias['Intervalo'][indice_modal].split(',')[0].strip('[](')) # Extraer límite inferior numérico
f_modal = tabla_frecuencias['Frec_Abs'][indice_modal]
f_prev = tabla_frecuencias['Frec_Abs'][indice_modal - 1] if indice_modal > 0 else 0
f_next = tabla_frecuencias['Frec_Abs'][indice_modal + 1] if indice_modal < len(tabla_frecuencias) - 1 else 0
# Calcular ancho del intervalo desde la tabla
h = float(tabla_frecuencias['Intervalo'][indice_modal].split(',')[1].strip(')[]')) - L_modal


# Evitar división por cero si los denominadores son cero
denominador_moda = (f_modal - f_prev) + (f_modal - f_next)
if denominador_moda != 0:
    moda = L_modal + ((f_modal - f_prev) / denominador_moda) * h
else:
    moda = "No se puede calcular (denominador cero)" # O manejar de otra manera


print(f"Moda Agrupad = L_modal + ((f_modal - f_prev) / (f_modal - f_prev) + (f_modal - f_next)) * {h:.1f} = {L_modal} + (({f_modal} - {f_prev}) / {denominador_moda}) * h = {moda:.2f}" if isinstance(moda, (int, float)) else f"Moda: {moda}")


# 3. Mediana (usando la tabla_frecuencias corregida)
N_mediana = total_datos / 2
clase_mediana = next(i for i, fa in enumerate(tabla_frecuencias['Frec_Abs_Acum']) if fa >= N_mediana)
L_mediana = float(tabla_frecuencias['Intervalo'][clase_mediana].split(',')[0].strip('[](')) # Extraer límite inferior numérico
f_mediana = tabla_frecuencias['Frec_Abs'][clase_mediana]
F_prev = tabla_frecuencias['Frec_Abs_Acum'][clase_mediana - 1] if clase_mediana > 0 else 0
# Calcular ancho del intervalo desde la tabla
h_mediana = float(tabla_frecuencias['Intervalo'][clase_mediana].split(',')[1].strip(')[]')) - L_mediana

# Evitar división por cero si la frecuencia mediana es cero
if f_mediana != 0:
    mediana = L_mediana + ((N_mediana - F_prev) / f_mediana) * h_mediana
else:
    mediana = "No se puede calcular (frecuencia mediana cero)" # O manejar de otra manera

print(f"Mediana Agrupada = L_mediana + ((N_mediana - F_prev) / f_mediana) * h_mediana = {L_mediana} + (({N_mediana} - {F_prev}) / {f_mediana}) * {h_mediana:.1f} =  {mediana:.2f}" if isinstance(mediana, (int, float)) else f"Mediana Agrupada: {mediana}")

```
